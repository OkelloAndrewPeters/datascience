```python
import pandas as pd
```

```python
data = pd.read_csv('iris.csv')
data
```

```python
#To see the species that are available
data['Species'].unique()
```

```python
#dealing with categorical data
data['Species'] = data['Species'].astype('category')
data['Species'] = data['Species'].cat.codes
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop(columns = ['Species'])
y = data['Species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 23)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(max_iter = 1000)
lr.fit(X_train, y_train)
```

```python
C = lr.intercept_
C
```

```python
m = lr.coef_
m
```

```python
y_pred_train = lr.predict(X_train)
```

```python
y_pred_train
```

```python
X_train
```

```python
lr.score(X_train, y_pred_train)
```

```python
y_pred_test = lr.predict(X_test)
```

```python
X_test
```

```python
y_pred_test
```

```python
lr.score(X_test, y_pred_test)
```

```python
predicted_output = lr.predict(X_test)
predicted_output
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, predicted_output)
cm
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize = (5, 4))
sns.heatmap(cm, annot = True)
plt.xlabel('Predicted Value')
plt.ylabel('Actual Value')
```

```python
#what the heatmap means
#the model has predicted zero where the actual value is zero and this has happened 15 times.
#the model has predicted one where the actual value is 1, and this has happened 8 times
#the model has predicted two where the actual value is 2 and this has happened 10 times
```
