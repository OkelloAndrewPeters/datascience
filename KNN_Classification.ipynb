```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
```

```python
df.head()
```

```python
df.shape
```

```python
#adding a new column in the same dataframe
df['target'] = iris.target
df.head()
```

```python
df[df.target == 1].head()
```

```python
df[df.target == 2].head()
```

```python
df[df.target==1].shape
```

```python
df[df.target==0].shape
```

```python
df['flower_name'] =df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
df[df.target==2].shape
```

```python
#spliting the dataframe into 3
df0 = df[:50]
df1 = df[50:100]
df2 = df[100:]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color = 'green', marker = '+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color = 'blue', marker = '.')
```

```python
#looking at the graph, we can clearly see that we can apply knn_classifier
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns = ['target', 'flower_name'])
y = df['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
#Creating a KNN Classifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
knn1 = KNeighborsClassifier(n_neighbors=10)
knn1.fit(X_train, y_train)
```

```python
knn1.score(X_test, y_test)
```

```python
#try to figure out best value of K, you can use grid search CV, or k fold cross validation
#let us use the confusion matrix to understand for which values it got the predictions right or wrong
```

```python
from sklearn.metrics import confusion_matrix

y_pred_test = knn1.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred_test)
cm
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize = (5, 4))
sns.heatmap(cm, annot = True)
plt.xlabel('Predicted Value')
plt.ylabel('Actual Value')
```

```python
#we can also use classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_test))
```
